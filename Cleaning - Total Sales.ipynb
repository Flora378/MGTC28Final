```python
# import pandas
import pandas as pd
```

```python
# load CSV data and declare row 0 as header
# https://pandas.pydata.org/docs/user_guide/advanced.html

df = pd.read_csv('D:\\Minda\\School\\University\\3 rd Year (2022-\\4 - Fall 2023 Semester\\MGTC28H3 - LEC 01\\Group Project - Nov 28\\data\\raw\\Total Sales - UTSC Lecture.csv', header=[0])
```

```python
df
```

```python
df.columns
```

```python
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.melt.html
df = df.melt(id_vars=['Fiscal Year', 'Fiscal Week'])
```

```python
df
```

```python
# rename the columns
df = df.rename(columns={df.columns[0]: "Year", df.columns[1]: "Week", df.columns[2]: "Need States", df.columns[3]: "Sales"})
```

```python
df
```

```python
# convert week# to an actual date for easy plotting
df['Date'] = pd.to_datetime(df["Year"].astype(str) + "-" + df["Week"].astype(str) + "-1", format='%Y-%U-%w')
```

```python
df
```

```python
import seaborn as sns
```

```python
# try a couple graphs
sns.lineplot(data = df, x=df.Date, y='Sales')
```

```python
sns.lineplot(data = df, x='Week', y='Sales', hue='Year')
```

## Additional:

```python
# rearrange columns
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reindex.html
# https://stackoverflow.com/questions/13148429/how-to-change-the-order-of-dataframe-columns
df = df.reindex(columns=['Year','Week','Date','Need States','Sales'])
```

```python
df
```

```python
# sort values so format matches other cleaned csv files
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html
df = df.sort_values(by=['Year', 'Week'])
```

```python
df
```

```python
# reset index numbers
df = df.reset_index(drop=True)
```

```python
df
```

```python
# replace instances of 0 with NaN
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html
# https://stackoverflow.com/questions/34794067/how-to-set-a-cell-to-nan-in-a-pandas-dataframe
df = df.replace(to_replace=0, value=pd.NA)
```

```python
df
```

```python
# remove rows that contain NaN
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html
df = df.dropna(axis=0)
```

```python
df
```

```python
# check how the graphs look now
sns.lineplot(data = df, x=df.Date, y='Sales')
```

```python
sns.lineplot(data = df, x='Week', y='Sales', hue='Year')
```

```python
# write the dataframe to a csv file
df.to_csv('D:\\Minda\\School\\University\\3 rd Year (2022-\\4 - Fall 2023 Semester\\MGTC28H3 - LEC 01\\Group Project - Nov 28\\data\\clean\\Total Sales.csv')
```
