```python
# First, let's import the library 'Pandas'
import pandas as pd
```

```python
# Loading the CSV data.  Noticed that I decleard row 0 as header.
# By doing this, I created a multi index. See documentation: https://pandas.pydata.org/docs/user_guide/advanced.html
df = pd.read_csv('/Users/katherinefeng/Desktop/kenvue_data/Total Trade Spend.csv')
```

```python
df
```

```python
# Let's rename the columns
df = df.rename_axis(None, axis=1)
df = df.rename(columns={df.columns[0]: "Year", df.columns[1]: "Week"})
```

```python
# Convert 0 in any columns to NaA
df[df == 0] = pd.NA
```

```python
# Drop any column that has NaN
df.dropna(inplace=True)
```
